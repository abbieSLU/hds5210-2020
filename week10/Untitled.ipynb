```python
import pandas as pd
from datetime import datetime
import csv
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
headers = ['Country','Code','Date','Johns Hopkins','WHO (cases)','ECDC (cases)']
df = pd.read_csv('covid-cases-by-source.csv',names=headers, parse_dates=['Date'])
mdf=df.fillna(0)
print (mdf)


xdf=df.groupby("Date")["Johns Hopkins"].count()
#only selected Johns Hopkins data seem more accurate 
y_new=xdf.cumsum()
xdf.plot()
```
